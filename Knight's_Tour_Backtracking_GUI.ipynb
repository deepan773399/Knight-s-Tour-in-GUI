```python
#!pip install turtle
import turtle 

sc = turtle.Screen()

pen = turtle.Turtle()

n=int(input("Enter the length of board  -"))
board=[[-1 for i in range(n)]for j in range(n)]

class knightproblem():   
    # create screen object
    

    def drawboard(self):

        sc.setup(600, 600)
        pen.speed(200)
        for i in range(n):
          pen.up()
          pen.setpos(0, 30 * i)
          pen.down()
          for j in range(n):
            for i in range(4):
                pen.forward(30)
                pen.left(90)

            pen.forward(30)


        pen.backward(30*n)
        pen.right(90)
        pen.forward(30*(n-1))
        pen.left(90)
        pen.up()


    def printsol(self):
        #It will print the moves/order in which the knight moves
        print("The knight moves in the following order --")
        for i in range(n):
            for j in range(n):
                print(board[i][j],end=" ")
            print("\n")

    def knightprob(self,curr_x,curr_y,move_x,move_y,pos):
        
        
        if pos==(n)*(n):
            #if total moves of knight is equal to total number of position of chess board than function will return true and knight will stop moving
            return True
        for i in range(8):
                #this loop is for total possibilities in which a knight can move which is 8
            newpos_x=curr_x+move_x[i]
            newpos_y=curr_y+move_y[i]
            if newpos_x>=0 and newpos_y>=0 and newpos_x<(n) and newpos_y<(n) and board[newpos_x][newpos_y]==-1:
                    #if knight is on chessboard and on a new postion than
                board[newpos_x][newpos_y]=pos+1
                if knightproblem.knightprob(self,newpos_x,newpos_y,move_x,move_y,pos+1) is True:
                        #This is recursion in which we move with curren position = new position and position = position+1
                    return True
                    #If at a point knight doesn't have any possibilty of moving than we will back on the intial position and initial coordinates and make the newly assigned positions as -1
                board[newpos_x][newpos_y]=-1
            
  
        return False  
 
    def drawpattern(self):
        pen.shape("triangle")
        pen.speed(1)
        for i in range(1,(n*n)+1):
            for j in board:
                if i in j:
                    pen.up()
                    pen.setposition(30*(board.index(j))+10,30*(j.index(i))+10)
                    pen.write(i)
                    pen.up()
                    
            




def main():
    move_x=[1,1,2,2,-1,-1,-2,-2]
    move_y=[2,-2,1,-1,2,-2,1,-1]
    pos=1
    board[0][0]=1
    curr_x=0
    curr_y=0
    a=knightproblem()
    
    a.drawboard()
    if a.knightprob(curr_x,curr_y,move_x,move_y,pos):
        a.drawpattern()
        a.printsol()
    
    else:
        print("Solution does not exist ")
    
 
if __name__=="__main__":
    main()



    
        
    
```
